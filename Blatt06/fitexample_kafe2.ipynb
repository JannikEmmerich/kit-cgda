# General example for fitting with kafe2
* read datafile fitexample.dat
* perform fit (2nd order polynomial)
* show and save output

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from kafe2 import XYFit, Plot, ContoursProfiler, XYContainer

# Define the model function for the fit

def poly2(x, a=1.0, b=0.0, c=0.0):
    return a * x**2 + b * x + c

# set data
x = [.05, 0.36, 0.68, 0.80, 1.09, 1.46, 1.71, 1.83, 2.44, 2.09, 3.72, 4.36, 4.60]
y = [0.35, 0.26, 0.52, 0.44, 0.48, 0.55, 0.66, 0.48, 0.75, 0.70, 0.75, 0.80, 0.90]
yerr = [0.06, 0.07, 0.05, 0.05, 0.07, 0.07, 0.09, 0.1, 0.11, 0.1, 0.11, 0.12, 0.1]

# Define the fit
fit = XYFit(xy_data = [x, y], model_function = poly2)

# Add errors
fit.add_error(axis = 'y', err_val = yerr)

# Optional: Assign LaTeX expressions to the function and its variables.
fit.assign_parameter_latex_names(a='A', b='B',c="C")
fit.assign_model_function_latex_expression("{a}\,x^2 + {b}\,x+{c}")

# Do the fit
fit.do_fit()

# Get a report about the fit in the output
fit.report()

# Finaly plot the fit
plot = Plot(fit)
plot.plot()

# create the contours profiler
#cpf = ContoursProfiler(fit)
#cpf.plot_profiles_contours_matrix()

# and show everything
plt.savefig("fitexample.pdf")
plt.show()
```

```python
data = np.loadtxt('fitexample_kafe2.txt').T
xy_data = XYContainer(data[0], data[1])
xy_data.add_error('y', 1)
```

# Aufgabe 6.2

```python
def linear(x, a0, a1):
    return a0 + a1 * x

fit2 = XYFit(xy_data, linear)
fit2.assign_parameter_latex_names(a0='a0', a1='a1')
fit2.assign_model_function_latex_expression("{a1}\,x+{a0}")

fit2.do_fit()
fit2.report()

plot = Plot(fit2)
plot.plot()
plt.show()
```

# Aufgabe 6.3

```python
cpf = ContoursProfiler(fit2)
cpf.plot_profiles_contours_matrix()
```

```python
def linear2(x, s, c):
    return s*(x - np.mean(xy_data.data[0])) + c

fit3 = XYFit(xy_data, linear2)
fit3.assign_parameter_latex_names(s='s', c='c')
fit3.assign_model_function_latex_expression("{s}\,(x - x_mean)+{c}")

fit3.do_fit()
fit3.report()

plt.close()
Plot(fit3).plot()
plt.show()
```

```python
cpf2 = ContoursProfiler(fit3)
cpf2.plot_profiles_contours_matrix()
```

Bei der Parametrisierung, so dass die Gerade um den Schwerpunkt der x-Werte dargestellt wird, sind s und c unkorreliert.

Die grafische Darstellung des Modells ist bei beiden Parametrisierung identisch. 
Die Unsicherheit der Steigung der Geraden ist bei beiden Varianten identisch.
Bei c handelt es sich allerdings um die Unsicherheit des Mittelwerts und bei a0 um die Unsicherheit des Ursprungs der Geraden.

Fr die Angabe meines Ergebnisses wrde ich die zweite Variante bevorzugen, da sie mir besser zu interpretierende Anpassungswerte liefert.
