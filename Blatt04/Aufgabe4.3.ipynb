## Teilaufgabe a

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import normaltest
from PhyPraKit import histstat
```

```python
µx, σx = 1.5, 0.5
µy, σy = 0.6, 0.15
```

```python
datax = np.random.normal(loc=µx, scale=σx, size=1000)
datay = np.random.normal(loc=µy, scale=σy, size=1000)
```

```python
v = datax + datay
plt.hist(v, bins=30)
plt.show()
```

## Teilaufgabe b

```python
σv = np.sqrt(σx**2 + σy**2)
µv = µx + µy
print("Standardabweichung der vi: " + str(σv))
quantiles = np.linspace(0, 4, 100)
plt.hist(v, density=True, bins=30)
plt.plot(quantiles, norm.pdf(quantiles, µv, σv))
plt.show()
```

## Teilaufgabe c

```python
w = datax / datay
```

```python
n, bins, _ = plt.hist(w, density=True, bins=30)
σw = np.sqrt(w.mean()**2 * ((σx / datax.mean())**2 + (σy / datay.mean())**2))
µw = µx / µy
print("Standardabweichung der wi: " + str(σw))
quantiles = np.linspace(bins[0], bins[-1], 100)
pdf = norm.pdf(quantiles, µw, σw)
plt.plot(quantiles, pdf)
plt.show()
```

```python
histstat(n, bins)
print("normal statistics:\n   mean=" + str(µw) + ", sigma=" + str(σw))
```
