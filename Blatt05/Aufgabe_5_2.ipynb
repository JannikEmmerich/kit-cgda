```python
import PhyPraKit as ppk
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = ppk.readCSV('signal_smooth.csv')[1]
```

## Teilaufgabe a

```python
autocorrelate = ppk.autocorrelate(data[1])
plt.plot(data[0], autocorrelate)
plt.xlabel('time (ms)')
plt.ylabel('autocorrelated amplitude (V)')
plt.show()
```

## Teilaufgabe b

```python
peakTimes = [data[0][i] for i in ppk.convolutionPeakfinder(autocorrelate)]
periods = [x - y for x, y in zip(peakTimes[1:], peakTimes)]
print('mean=' + str(np.mean(periods)) + '\tstd=' + str(np.std(periods)))
```
