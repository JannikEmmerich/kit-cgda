```python
import PhyPraKit as ppk
import matplotlib.pyplot as plt
```

## Teilaufgabe a

```python
data = ppk.readCSV('signal_raw.csv')[1]
plt.plot(data[0], data[1])
plt.xlabel('time (ms)')
plt.ylabel('amplitude (V)')
plt.show()
```

## Teilaufgabe b

```python
smoothData = ppk.meanFilter(data[1], 7)
plt.plot(data[0], smoothData)
plt.xlabel('time (ms)')
plt.ylabel('smoothed amplitude (V)')
plt.show()
```

```python
diffData = smoothData - data[1]
plt.plot(data[0], diffData)
plt.xlabel('time (ms)')
plt.ylabel('diff amplitude (V)')
plt.show()
```

Wenn man den Glättungsparameter zu klein wählt, ist die Kurve nach dem glätten noch sehr stark verrauscht.
Wählt man den Glättungsparameter zu groß, so gehen eventuell wichtige Details der Kurve verloren, da sie sozusagen "weggeglättet" werden.
