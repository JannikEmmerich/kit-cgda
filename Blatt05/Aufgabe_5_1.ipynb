```python
import PhyPraKit as ppk
import matplotlib.pyplot as plt
```

## Teilaufgabe a

```python
data = ppk.readCSV('signal_raw.csv')[1]
plt.plot(data[0], data[1])
plt.xlabel('time (ms)')
plt.ylabel('amplitude (V)')
plt.show()
```

## Teilaufgabe b

```python
smoothData = ppk.meanFilter(data[1], 7)
plt.plot(data[0], smoothData)
plt.xlabel('time (ms)')
plt.ylabel('smoothed amplitude (V)')
plt.show()
```

```python
diffData = smoothData - data[1]
plt.plot(data[0], diffData)
plt.xlabel('time (ms)')
plt.ylabel('diff amplitude (V)')
plt.show()
```
