```python
import PhyPraKit as ppk
import matplotlib.pyplot as plt
```

```python
data = ppk.readCSV('signal_smooth.csv')[1]
```

```python
freq, amp = ppk.FourierSpectrum(data[0], data[1])
plt.plot(freq, amp)
plt.xlabel('frequency (1/s)')
plt.ylabel('amplitude (V)')
plt.show()
```

Das Signal setzt sich aus einer Frequenz mit sehr großer Amplitude ($f_1 \approx 0.3 \frac{1}{s}; A_1 \approx 0.5 V$) und einer Frequenz mit etwas kleinerer Amplitude ($f_2 \approx 0.6 \frac{1}{s}; A_2 \approx 0.1 V$) zusammen. Das dritte kleine lokale Maximum ist entweder als Rauschen zu erklären oder es handelt sich dabei um einer dritte Frequenz im Spektrum.
