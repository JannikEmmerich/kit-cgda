```python
import PhyPraKit as ppk
import matplotlib.pyplot as plt
```

```python
data = ppk.readCSV('signal.csv')[1]
```

```python
plt.plot(data[0], data[1])
plt.xlabel('Time (ms)')
plt.ylabel('Recording (a.u.)')
plt.show()
```

```python
freq, amp = ppk.FourierSpectrum(data[0], data[1])
```

```python
plt.plot(freq, amp)
plt.xlim(0, 6)
plt.xlabel('frequency (1/ms)')
plt.ylabel('amplitude (a.u.)')
plt.show()
```

```python
filteredData = ppk.meanFilter(data[1], width=25)
plt.plot(data[0], filteredData)
plt.xlabel('Time (ms)')
plt.ylabel('Recording (a.u.)')
plt.show()
```

```python
freq_, amp_ = ppk.FourierSpectrum(data[0], filteredData)
```

```python
plt.plot(freq_, amp_)
plt.xlim(0, 6)
plt.xlabel('frequency (1/ms)')
plt.ylabel('amplitude (a.u.)')
plt.show()
```
