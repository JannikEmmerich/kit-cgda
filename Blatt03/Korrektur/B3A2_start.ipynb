```python
# Aufgabenteil b
N = 1000

def wuerfele(n):
    S = 0
    for i in range(0, n):
        S += np.random.randint(1, 7)
    return S / n

# n = 10
augenzahlen = np.zeros(N, dtype=float)
for i in range(0, N):
    augenzahlen[i] = wuerfele(10)
plt.hist(augenzahlen)
plt.xlabel("Augenzahl")
plt.ylabel("Häufigkeit")
plt.title("n = 10")
plt.show()

# n = 50
augenzahlen = np.zeros(N, dtype=float)
for i in range(0, N):
    augenzahlen[i] = wuerfele(50)
plt.hist(augenzahlen)
plt.xlabel("Augenzahl")
plt.ylabel("Häufigkeit")
plt.title("n = 50")
plt.show()

# n = 100
augenzahlen = np.zeros(N, dtype=float)
for i in range(0, N):
    augenzahlen[i] = wuerfele(100)
n, bins, patches = plt.hist(augenzahlen)
plt.xlabel("Augenzahl")
plt.ylabel("Häufigkeit")
plt.title("n = 100")
plt.show()
```

```python
# Aufgabenteil c
import scipy.stats as stats

augenzahlen = np.zeros(N, dtype=float)
for i in range(0, N):
    augenzahlen[i] = wuerfele(100)
n, bins, patches = plt.hist(augenzahlen)

a = np.array(n)
b = np.delete(bins, -1)

mean = sum(a*b) / sum(a)

sumV = 0
for i in range(0, len(a)):
       sumV += a[i] * (b[i] - mean)**2
variance = sumV / sum(a)

sigma = variance**(1/2)
mu = mean  
x = np.linspace(0, 6, 1000)
plt.plot(x, np.amax(a) * 1/2 * stats.norm.pdf(x, mu, sigma))

plt.xlabel("Augenzahl")
plt.ylabel("Häufigkeit")
plt.title("n = 100")
plt.show()

# Beobachtung: Die mittleren Augenzahlen sind normalverteilt.
```
