```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rand = np.random
```

```python
x1 = np.arange(0, 5, 0.01)
y1 = x1**2
plt.plot(x1, y1)

x2 = np.array([1, 2, 3, 4])
y2 = rand.normal(size=4) * 0.1 * (x2**2) + x2**2
plt.errorbar(x2, y2, 0.1*(x2**2), fmt='bo')

plt.show()
```
